```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
url = 'https://www.klimaatinfo.nl/verenigde-staten-van-amerika/arizona/yuma.htm'

html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")

table = soup.find_all('table', {'id': 'AutoNumber3'})[0]

rows = table.find_all('tr')
rows

rows_2 = [row.find_all('td') for row in rows]


text = [[x.text for x in row] for row in rows_2]

colnames = ['month', 'avg_max_temp', 'avg_min_temp', 'hours_of_sun', 'rain_days_per_month', 'avg_rain', 'avg_sea_temp']

climate_raw = pd.DataFrame(text, columns = colnames)

climate_raw_2 = climate_raw.drop(climate_raw.index[[0,1,14]])
climate_raw_2

del climate_raw_2['avg_rain']
climate_raw_2


#climate_raw['month'] = climate_raw['month'].map(lambda x: x.lstrip('\n'))

#climate_raw['avg max temp'] = climate_raw['avg max temp'].map(lambda x: x.lstrip('\r\r\n'))
```

```python
climate_raw_2['month'] = climate_raw_2['month'].map(lambda x: x.lstrip('\n'))
climate_raw_2['avg_max_temp'] = climate_raw_2['avg_max_temp'].map(lambda x: x.lstrip('\r\r\n'))
climate_raw_2['avg_min_temp'] = climate_raw_2['avg_min_temp'].map(lambda x: x.lstrip('\r\r\n'))
climate_raw_2['hours_of_sun'] = climate_raw_2['hours_of_sun'].map(lambda x: x.lstrip('\r\r\n'))
climate_raw_2['rain_days_per_month'] = climate_raw_2['rain_days_per_month'].map(lambda x: x.lstrip('\r\r\n'))
climate_raw_2['avg_sea_temp'] = climate_raw_2['avg_sea_temp'].map(lambda x: x.lstrip('\r\r\n'))
climate_raw_2
```

```python
climate_raw_2.to_csv('Mexicali.csv')
```
